```python
from scipy.stats import pearsonr

corr, p = pearsonr(merged['AUC'], merged['Z_SCORE'])
print(f"Pearson correlation: {corr:.2f} (p={p:.2e})")
```

```python
# Defining thresholds
q25 = merged['AUC'].quantile(0.25)
q75 = merged['AUC'].quantile(0.75)

# Classifying response based on AUC
def classify_auc(x):
    if x <= q25:
        return 'Sensitive'
    elif x >= q75:
        return 'Resistant'
    else:
        return 'Intermediate'

merged['response'] = merged['AUC'].apply(classify_auc)
```

```python
# Filtering only the two extreme groups
binary_df = merged[merged['response'].isin(['Sensitive', 'Resistant'])]
```

```python
from scipy.stats import ttest_ind

# Keep only gene columns (exclude metadata)
gene_cols = merged.columns[1:-41]  # Last 41 columns are metadata

# Separating expression data
sensitive_expr = binary_df[binary_df['response'] == 'Sensitive'][gene_cols]
resistant_expr = binary_df[binary_df['response'] == 'Resistant'][gene_cols]

# Converting gene expression values to numeric (in place), coerce errors
sensitive_expr = sensitive_expr.apply(pd.to_numeric, errors='coerce')
resistant_expr = resistant_expr.apply(pd.to_numeric, errors='coerce')

# Running t-test per gene
import pandas as pd
results = []
for gene in gene_cols:
    t_stat, p_val = ttest_ind(sensitive_expr[gene], resistant_expr[gene], equal_var=False)
    log2_fc = sensitive_expr[gene].mean() - resistant_expr[gene].mean()
    results.append({'Gene': gene, 'p_value': p_val, 'log2_FC': log2_fc})

results_df = pd.DataFrame(results)
results_df['adjusted_p'] = results_df['p_value'] * len(results_df)  # Bonferroni

```

```python
# Genes with significant adjusted p-values (Bonferroni < 0.05)
significant_genes = results_df[results_df['adjusted_p'] < 0.05]
```

```python
import matplotlib.pyplot as plt
import numpy as np

results_df['-log10_p'] = -np.log10(results_df['p_value'])

plt.figure(figsize=(10, 6))
plt.scatter(results_df['log2_FC'], results_df['-log10_p'], 
            c=(results_df['adjusted_p'] < 0.05), cmap='coolwarm', alpha=0.7)

plt.axhline(-np.log10(0.05), color='grey', linestyle='--')
plt.axvline(1, color='grey', linestyle='--')
plt.axvline(-1, color='grey', linestyle='--')
plt.xlabel('Log2 Fold Change')
plt.ylabel('-Log10 p-value')
plt.title('Volcano Plot: Sensitive vs Resistant (AUC-based)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests

# Defining gene expression columns
gene_cols = merged.columns[1:-10]  # Adjust this if necessary based on column position

# Splitting into groups
sensitive = merged[merged['response'] == 'Sensitive']
resistant = merged[merged['response'] == 'Resistant']
```

```python
# Convert gene expression columns to numeric
merged[gene_cols] = merged[gene_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Mann–Whitney U test per gene
results = []
for gene in gene_cols:
    x = sensitive[gene].dropna()
    y = resistant[gene].dropna()
    if len(x) > 0 and len(y) > 0:
        stat, p = mannwhitneyu(x, y, alternative='two-sided')
        log2fc = np.log2(x.mean() + 1e-6) - np.log2(y.mean() + 1e-6)
        results.append({'gene': gene, 'log2FC': log2fc, 'p_value': p})

# Adjusting p-values
results_df = pd.DataFrame(results)
results_df['adj_p_value'] = multipletests(results_df['p_value'], method='fdr_bh')[1]

results_df = results_df.sort_values('adj_p_value')
results_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Preparing volcano plot
df_volcano = results_df.copy()
df_volcano['-log10(adj_p_value)'] = -np.log10(df_volcano['adj_p_value'])

top_hits = df_volcano.sort_values('adj_p_value').head(5)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_volcano, x='log2FC', y='-log10(adj_p_value)', hue=df_volcano['adj_p_value'] < 0.05, palette={True: 'red', False: 'gray'})

plt.axhline(-np.log10(0.05), color='blue', linestyle='--')
plt.xlabel('Log2 Fold Change')
plt.ylabel('-Log10 Adjusted P-value')
for _, row in top_hits.iterrows():
    plt.text(row['log2FC'], row['-log10(adj_p_value)'], row['gene'], fontsize=8)
plt.title('Volcano Plot of Differential Expression between Lapatinib-Sensitive and -Resistant Cell Lines')
plt.legend(title='Significant (FDR < 0.05)')
plt.tight_layout()
plt.show()
```
