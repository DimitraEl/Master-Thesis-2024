```python
lapatinib_data = drug_sensitivity[drug_sensitivity['DRUG_NAME'] == 'Lapatinib']

print(lapatinib_data.head())
print(f"Number of cell lines tested with Lapatinib: {len(lapatinib_data)}")
```

```python
ln_ic50 = lapatinib_data['LN_IC50']

print(ln_ic50.describe())
```

```python
print(f"Missing LN_IC50 values: {ln_ic50.isna().sum()}")
lapatinib_data_cleaned = lapatinib_data.dropna(subset=['LN_IC50'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.histplot(ln_ic50, kde=True, bins=30, color="blue")
plt.title("Distribution of Lapatinib Sensitivity (LN_IC50)")
plt.xlabel("LN_IC50")
plt.ylabel("Frequency")
plt.show()
plt.savefig("Distribution of Lapatinib Sensitivity")
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=ln_ic50, color="skyblue")
plt.title("Boxplot of Lapatinib Sensitivity (LN_IC50)")
plt.xlabel("LN_IC50")
plt.savefig("boxplot of lap sensitivity")
plt.show()
```

```python
merged_data = pd.merge(metadata, lapatinib_data, how='inner', left_on='Sanger_Model_ID', right_on='SANGER_MODEL_ID')

tissue_sensitivity = merged_data.groupby('lineage')['LN_IC50'].mean()

plt.figure(figsize=(12, 6))
tissue_sensitivity.sort_values().plot(kind="bar", color="coral")
plt.title("Average Lapatinib Sensitivity by Tissue Type")
plt.ylabel("Mean LN_IC50")
plt.xlabel("Tissue Type")
plt.xticks(rotation=45)
plt.savefig("Average Lap Sensitivity by Tissue Type")
plt.show()
```

```python
# Identify highly sensitive cell lines
low_sensitivity = lapatinib_data[lapatinib_data['LN_IC50'] < ln_ic50.quantile(0.05)]
print("Highly sensitive cell lines:")
print(low_sensitivity[['SANGER_MODEL_ID', 'LN_IC50']])

# Identify highly resistant cell lines
high_resistance = lapatinib_data[lapatinib_data['LN_IC50'] > ln_ic50.quantile(0.95)]
print("Highly resistant cell lines:")
print(high_resistance[['SANGER_MODEL_ID', 'LN_IC50']])
```

```python
# Data Cleaning: handle missing or incorrect values
plot_df = merged_expression_sensitivity[['LN_IC50', 'sex', 'age', 'lineage']].dropna()
plot_df = plot_df[plot_df['sex'].isin(['Male', 'Female'])]

# 1. Gender Comparison
plt.figure(figsize=(8, 6))
if not plot_df.empty:
    sns.boxplot(x='sex', y='LN_IC50', data=plot_df)
    plt.title('Lapatinib Sensitivity by Gender')
    plt.savefig("Lapatinib Sensitivity by Gender.png")
    plt.show()
else:
    print("No data to plot for Gender Comparison.")

# 2. Age Group Analysis
plot_df['age_group'] = pd.cut(plot_df['age'], bins=[0, 30, 50, 70, 100], labels=['<30', '30-50', '51-70', '70+'])
plt.figure(figsize=(10, 6))
if not plot_df.empty:
    sns.violinplot(x='age_group', y='LN_IC50', data=plot_df)
    plt.title('Lapatinib Sensitivity by Age Group')
    plt.savefig("Lapatinib Sensitivity by Age Group.png")
    plt.show()
else:
    print("No data to plot for Age Group Analysis.")

# 3. Cancer Type
plt.figure(figsize=(12, 6))
if not plot_df.empty:
    cancer_order = plot_df.groupby('lineage')['LN_IC50'].median().sort_values().index
    sns.barplot(x='lineage', y='LN_IC50', data=plot_df, order=cancer_order)
    plt.xticks(rotation=90)
    plt.title('Lapatinib Sensitivity by Cancer Type')
    plt.tight_layout()
    plt.savefig("Lapatinib Sensitivity by Cancer Type.png")
    plt.show()
else:
    print("No data to plot for Cancer Type.")
```
